## Missing values Treatment

```python
Automobile_data.replace('?',np.nan, inplace=True)
```

```python
Automobile_data[['normalized-losses','bore','stroke','horsepower','peak-rpm','price']] = Automobile_data[['normalized-losses','bore','stroke','horsepower','peak-rpm','price']].astype('float64')
```

```python
Automobile_data.isnull().sum()
```

__Observations__

From above we see __normalized-losses__ have 41 null values, __num-of-doors__ have 2 null values, __bore,stroke and price__ have 4 null values and __horsepower and peak-rpm__ have 2 null values in each column.

```python
Automobile_data['normalized-losses'].fillna(Automobile_data['normalized-losses'].mean(), inplace=True)
```

```python
Automobile_data['bore'].fillna(Automobile_data['bore'].mean(), inplace=True)
```

```python
Automobile_data['stroke'].fillna(Automobile_data['stroke'].mean(), inplace=True)
```

```python
Automobile_data['horsepower'].fillna(Automobile_data['horsepower'].mean(), inplace=True)
```

```python
Automobile_data['peak-rpm'].fillna(Automobile_data['peak-rpm'].mean(), inplace=True)
```

```python
Automobile_data['price'].fillna(Automobile_data['price'].mean(), inplace=True)
```

```python
Automobile_data['num-of-doors'].fillna('four', inplace=True)
```

```python
Automobile_cat = Automobile_data.select_dtypes(include=['object'])
Automobile_cat.columns.tolist()
```

```python
for i in Automobile_cat.columns.tolist():
    
    dist= pd.DataFrame(Automobile_data[i].value_counts().sort_values(ascending=False)).reset_index()

    ax = dist.plot.bar(x='index',y=i,fontsize=9,color='maroon')

    ax.set_title("Distribution of "+i)
    ax.set_xlabel(i)

    #plt.savefig('C:/project/Count plot for '+i+'.png', bbox_inches='tight')
```

__Observations__

1)From the above 1st bar graphs we observe the __Top 5__ common cars in make are __Toyota,Nissan,Mazda,Honda and Mitsubishi__ which covers   60% of all the cars.

2)In the 2nd bar graph we observe more manufactures of __gas cars__ compared to __diesel cars__ .

3)In the 4th bar graph we see cars with __four num-of-doors__ are manufactured the most in comparison to the __two doors__.

4)5th graph shows us cars with __Top 3__ body-styles are __sedan,hatch-back and wagon__ leading the most.

5)The 6th graph shows us cars with ___drive wheels fwd is highly manufactured__ i.e 50% of all the cars where as __4wd drive       wheels__ are seen at the lowest i.e 5% .

6)In the 7th graph it can be seen 90% of the cars have the front engine location.

7)Cars with Top most engine type is seen ohc i.e approx 80% of all the engine-type.

8)Also 80% cars  with 4 cylinders are seen manufacturing the most where 12 and 3 cylinder cars are seen to be rare.

```python
for i in (Automobile_data.select_dtypes(include=['object'])).columns.tolist():
    
    price_by_make= Automobile_data.groupby([i])['price'].mean().sort_values(ascending=False).reset_index()
    

    ax = price_by_make.plot.bar(x=i,y="price",fontsize=9)
   
    ax.set_title("Average price by "+i)
    ax.set_ylabel('Average Price')
    ax.set_xlabel(i)

   # plt.savefig('C:/project/Average price by '+i+'.png', bbox_inches='tight')
```
